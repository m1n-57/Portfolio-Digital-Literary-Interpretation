추가 코드 : 계산되어 나온 Significant Fluctuation Point의 Index 원문을 직접 보고 싶다면, 아래의 코드를 이용하면 됩니다.

```python
import re

# 파일 경로 설정
file_path = '/content/sample_data/Pride-and-Prejudice.txt'  # 파일 경로를 필요에 맞게 수정한다.

# 텍스트 파일 읽기
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# 텍스트를 문단 단위로 분할 (빈 줄과 장 제목을 기준으로)
paragraphs = re.split(r'\n\n|Chapter \d+\n', text, flags=re.IGNORECASE)
paragraphs = [paragraph.strip() for paragraph in paragraphs if paragraph.strip()]  # 빈 문단 제거 및 공백 제거

# 특정 인덱스 범위 설정
index_ranges_to_print = [
    (170, 282),
    (735, 847),
    (1187, 1299)
]  # 여기에 출력하고 싶은 인덱스 범위를 추가

# 인덱스 범위 출력
for start_idx, end_idx in index_ranges_to_print:
    if 0 <= start_idx < len(paragraphs) and 0 <= end_idx < len(paragraphs):
        print(f"Index range {start_idx} to {end_idx}:")
        for idx in range(start_idx, end_idx + 1):
            print(f"\n[Paragraph {idx}] {paragraphs[idx]}")
        print("\n" + "="*50 + "\n")
    else:
        print(f"Index range {start_idx} to {end_idx} is out of range.")
```
